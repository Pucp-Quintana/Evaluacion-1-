```python
#1 Subir la data y limpiar
#Emisiones de CO2
from IPython.display import IFrame  
emisiones= "https://edgar.jrc.ec.europa.eu/report_2022#emissions_table"
```

```python
 import pandas as pd

emsions=pd.read_html(emisiones, # link
                        header=0, # where is the header?
                        flavor='bs4')
emisions
```

```python
emisions1 = emisions.transpose() 
emisions1
```

```python
emisions1.info() 
```

```python
#Utiice  str.replace para cambiar la , a . 
emisions2 = emisions1.apply(lambda x: x.str.replace(',', '.', regex=True))
emisions2
```

```python
#Utilice la funcion de strip() y .applymap() para eliminar los espacios en blanco de los costados
emisions3 = emisions2.applymap(lambda x: x.strip() if isinstance(x, str) else x)
emisions3
```

```python
emisions4 = emisions3.applymap(lambda x: x.strip() if isinstance(x, str) else x)
emisions4
```

```python
#Usamos la funcion .replace para eliminar los espacios en blanco 
emisions5 = emisions4.replace(r'\s+', '', regex=True)
emisions5
```
