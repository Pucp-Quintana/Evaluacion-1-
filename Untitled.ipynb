```python
import pandas as pd
from IPython.display import IFrame

# Emisiones de CO2
emisioneslink = "https://edgar.jrc.ec.europa.eu/report_2022#emissions_table"
emisiones = pd.read_html(emisioneslink,header=0,flavor="bs4")

# Temperatura mundial
tempelink = "https://www.worldeconomics.com/Indicator-Data/ESG/Environment/Temperatures/" 
tempe = pd.read_html(tempelink, header=0,flavor="bs4")

# Tasa de crecimiento de la produccion industrial
tilink = "https://www.cia.gov/the-world-factbook/field/industrial-production-growth-rate/country-comparison/"
ti = pd.read_html(tilink,header=0,flavor="bs4")
```

```python
#Con este codigo empleado set encontraremos la diferencia entre dos conjuntos de paises, los paises en emisiones1 pero no en ti1
Onlyemisions= set(emisiones1.Country)-set(ti1.Country)
Onlyemisions
```

```python
#Con este codigo empleado set encontraremos la diferencia entre dos conjuntos de paises, los paises en ti1 pero no en emisiones1
Onlyti= set(ti1.Country)-set(emisiones1.Country)
Onlyti
```

```python
#La funcion extractOne permite buscar elementos similares en dos listas
from thefuzz import process as fz
[(f,fz.extractOne(f,Onlyemisions)) for f in sorted(Onlyti)]
```

```python
#Construiremos un diccionario de coincidencias que superan un cierto umbral de similitud o coincidencia (en este caso, 87).
#entre las dos variables
[(f,fz.extractOne(f ,Onlyemisions)) for f in sorted(Onlyti)
 if fz.extractOne(f ,Onlyemisions)[1]>=87]
```

```python
#La variable cambioenemisionytierá un diccionario que contiene las coincidencias encontradas entre elementos de Onlyti y Onlyemisions 
#que cumplen con la condición de tener una similitud igual o superior a 87.
cambioenemisionyti = {f:fz.extractOne(f, Onlyemisions)[0] 
                 for f in sorted(Onlyti)
                 if fz.extractOne(f, Onlyemisions)[1] >=87}
cambioenemisionyti
```

```python
#Se usa el replace() para reemplazar el diccionario creado con los datos similares
emisiones1['Country'].replace(cambioenemisionyti, inplace=True)
emisiones1
```
